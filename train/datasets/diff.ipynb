```python
import numpy as np
from PIL import Image
import nibabel as nib
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
from data_preV2 import get_data_path, load_data
DATA_DIR = "./datasets"
IMG_TYPE = ("flair", "t1", "t1ce", "t2")
MASK_TYPE = "seg"
PATH_ID = 1
KEY_FRAME = 60
THRESHOLD = 0.75

path = get_data_path(Path(DATA_DIR))[PATH_ID]
print(f"Path: {path}, frame: {KEY_FRAME}")
images = load_data(path, "img", KEY_FRAME)
true_mask = load_data(path, "mask", KEY_FRAME)
raw_image = images[0]

plt.imshow(raw_image)
plt.show()
```

```python
png_image = Image.open("./nii2jpg_output/BraTS2021_00621/BraTS2021_00621_flair.nii.png")
png_image = np.array(png_image) / 255
jpg_image = Image.open("./nii2jpg_output2/BraTS2021_00621/BraTS2021_00621_flair.nii.jpeg")
jpg_image = np.array(jpg_image) / 255

diff1 = png_image - raw_image
diff2 = jpg_image - raw_image
print(diff1.max(), diff1.min(), diff1.mean())
print(diff2.max(), diff2.min(), diff2.mean())

plt.subplot(1, 2, 1).imshow(diff1)
plt.title("png")
plt.subplot(1, 2, 2).imshow(diff2)
plt.title("jpg")
plt.show()
```
